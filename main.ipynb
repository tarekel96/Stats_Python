```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/all_seasons.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
df = dataset.iloc[:, 1:]
```

```python
df.head()
```

```python
df.info()
```

```python
age, height, pts = df['age'].values, df['player_height'].values, df['pts'].values
```

```python
def plot_2d(x, y):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.show()
```

```python
plot_2d(age, pts)
```

```python
len(age), len(height)
```

```python
df_agg = df[['age', 'pts']].groupby('age', as_index=False).agg('mean')
df_agg
```

```python
df_agg.columns
```

```python
agg_age, agg_pts = df_agg['age'].values, df_agg['pts'].values
```

```python
plot_2d(agg_age, agg_pts)
```
